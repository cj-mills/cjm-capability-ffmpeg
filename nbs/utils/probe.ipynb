# Media Probing

> Probe media files for metadata (duration, ...) via ffprobe.

```python
#| export
from pathlib import Path
from typing import Optional
import subprocess

def get_media_duration(file_path: Path  # Path to the media file
                      ) -> Optional[float]:  # Duration in seconds, or None if it cannot be determined
    """Get the duration of a media file (seconds) via ffprobe."""
    duration_cmd = [
        'ffprobe',
        '-v', 'quiet',
        '-show_entries', 'format=duration',
        '-of', 'csv=p=0',
        str(file_path),
    ]
    try:
        duration_result = subprocess.run(
            duration_cmd, check=True, capture_output=True, text=True,
        )
        return float(duration_result.stdout.strip())
    except (subprocess.CalledProcessError, ValueError, FileNotFoundError):
        return None
```

```python
# Non-existent file -> None (ffprobe fails, swallowed)
assert get_media_duration(Path("/nonexistent/file.mp3")) is None
print("probe OK")
```
